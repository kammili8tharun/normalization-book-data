```python
# Create dataframe from dictionary
df = pd.DataFrame(data)
df
```

# First Normal Form (1NF)

```python
# 1. Each row must be unique
if df.duplicated().any() == False :
    print("Rows are unique")
else:
    print("Rows are nto unique")
```

```python
# 2. Each cell must contain unique value
df
```

# Second normal form (2NF)

```python
# 1.Database is in first normal form (1N)
df
```

```python
# 2. Seperate groups of values that apply to multiple rows by creating new tables

# In above table 'AUTHOR_NAME', 'AUTHOR_BIO' apply to multilple rows, so seperating to different rows

BookShop_AuthorDetails = df[['AUTHOR_ID', 'AUTHOR_NAME', 'AUTHOR_BIO']]
BookShop_AuthorDetails
```

```python
# Modifying the BookShop dataframe to drop the redundant data in columns 'AUTHOR_NAME', 'AUTHOR_BIO' that formed new table

BookShop = df
BookShop = BookShop.drop(columns=['AUTHOR_NAME', 'AUTHOR_BIO'])
```

```python
BookShop_AuthorDetails = BookShop_AuthorDetails.drop_duplicates(ignore_index=True).sort_values(by = ['AUTHOR_ID']).reset_index(drop = True)
```

# Load normalised tables to database train

```python
def load_to_db(df, sql_connection, table_name):
    df.to_sql(table_name, sql_connection, if_exists='replace', index=False)
```

```python
import pymysql
from sqlalchemy import create_engine
import mysql.connector

sql_connection = create_engine('mysql+pymysql://root:12345678@localhost:3306/train')

load_to_db(BookShop, sql_connection, 'BookShop')

sql_connection.dispose()
```

```python
sql_connection = create_engine('mysql+pymysql://root:12345678@localhost:3306/train')

load_to_db(BookShop_AuthorDetails, sql_connection, 'BookShop_AuthorDetails')

sql_connection.dispose()
```
